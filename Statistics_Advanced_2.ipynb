```python
'''5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, test='right'):
    """
    Visualizes Type I and Type II errors for one-tailed Z-test.

    Parameters:
    - mu0: Mean under null hypothesis
    - mu1: Mean under alternative hypothesis
    - sigma: Standard deviation
    - alpha: Significance level (Type I error)
    - test: 'right' or 'left' for one-tailed tests
    """

    # Create x-axis range
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    # Null and alternative distributions
    y0 = norm.pdf(x, mu0, sigma)  # Hâ‚€
    y1 = norm.pdf(x, mu1, sigma)  # Hâ‚

    # Determine critical value and beta region
    if test == 'right':
        z_critical = norm.ppf(1 - alpha)
        critical_val = mu0 + z_critical * sigma
        type1_region = x > critical_val
        type2_region = x < critical_val
    else:
        z_critical = norm.ppf(alpha)
        critical_val = mu0 + z_critical * sigma
        type1_region = x < critical_val
        type2_region = x > critical_val

    # Calculate Type II error (Î²)
    beta = norm.cdf(critical_val, mu1, sigma) if test == 'right' else 1 - norm.cdf(critical_val, mu1, sigma)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='Hâ‚€ Distribution', color='black')
    plt.plot(x, y1, label='Hâ‚ Distribution', color='green')

    # Shade Type I error (alpha)
    plt.fill_between(x, y0, where=type1_region, color='red', alpha=0.4, label='Type I Error (Î±)')

    # Shade Type II error (beta)
    plt.fill_between(x, y1, where=type2_region, color='blue', alpha=0.3, label='Type II Error (Î²)')

    # Annotations
    plt.axvline(critical_val, color='purple', linestyle='--', label=f"Critical Value = {critical_val:.2f}")
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Print error values
    print(f"Type I Error (Î±): {alpha:.3f}")
    print(f"Type II Error (Î²): {beta:.3f}")
    print(f"Power of the test (1 - Î²): {1 - beta:.3f}")
```

```python
'''16.Implement an F-test using Python to compare the variances of two random samples'''
import numpy as np
from scipy.stats import f
np.random.seed(34)
group_1=np.random.normal(170,5,25)
group_2=np.random.normal(165,4,23)
group_1_var=np.var(group_1)
group_2_var=np.var(group_2)
alpha=.05;
dof1=len(group_1)-1
dof2=len(group_2)-1
if (group_1_var>group_2_var):
  F_test=group_1_var/group_2_var
else:
  F_test=group_2_var/group_1_var
F_critical=f.ppf(alpha,dof1,dof2)
if(F_test>F_critical):
  print("reject the null hypotheis")
else:
  print("fail to reject the null hypothesis")
```

```python
'''19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA'''
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    print("----- Checking ANOVA Assumptions -----")

    # 1. Normality (Shapiro-Wilk Test for each group)
    print("\n1. Normality Test (Shapiro-Wilk):")
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        print(f"  Group {i}: W={stat:.4f}, p-value={p:.4f} ->",
              "Normal" if p > 0.05 else "Not normal")

    # 2. Equal Variances (Levene's Test)
    print("\n2. Homogeneity of Variances (Levene's Test):")
    stat, p = levene(*groups)
    print(f"  Levene Stat={stat:.4f}, p-value={p:.4f} ->",
          "Equal variances" if p > 0.05 else "Unequal variances")

    # 3. Independence
    print("\n3. Independence:")
    print("  Independence should be ensured by study design (random, independent samples).")
    print("  This cannot be statistically tested directly from one-way ANOVA sample data.")

    print("\n--------------------------------------")
```

```python
'''20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results'''
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a sample dataset
data = pd.DataFrame({
    'Score': [88, 92, 85, 90, 86, 79, 83, 80, 84, 77, 89, 87],
    'Teaching_Method': ['Online', 'Online', 'Online', 'Online', 'Offline', 'Offline', 'Offline', 'Offline', 'Hybrid', 'Hybrid', 'Hybrid', 'Hybrid'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']
})

# Step 2: Two-way ANOVA with interaction
model = ols('Score ~ C(Teaching_Method) * C(Gender)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("Two-Way ANOVA Results:\n")
print(anova_table)

# Step 3: Visualize interaction using seaborn
plt.figure(figsize=(8, 5))
sns.pointplot(data=data, x='Teaching_Method', y='Score', hue='Gender', dodge=True, markers=['o', 's'], capsize=0.1)
plt.title("Interaction Plot: Teaching Method vs Gender")
plt.ylabel("Mean Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means '''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate data for 3 groups
np.random.seed(0)
group1 = np.random.normal(20, 5, 30)
group2 = np.random.normal(22, 5, 30)
group3 = np.random.normal(24, 5, 30)

# ANOVA test
f_stat, p_value = stats.f_oneway(group1, group2, group3)
print("F-statistic:", round(f_stat, 3))
print("P-value:", round(p_value, 4))

# Boxplot to visualize
plt.boxplot([group1, group2, group3], labels=["Group 1", "Group 2", "Group 3"])
plt.title("One-Way ANOVA: Group Comparison")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
'''24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results '''
# Simulated data
sample = np.random.normal(50, 8, 30)
sample_var = np.var(sample, ddof=1)
pop_var = 64  # Hypothesized population variance (8^2)
n = len(sample)

# Chi-square test
chi2_stat = (n - 1) * sample_var / pop_var
p_value = stats.chi2.sf(chi2_stat, df=n-1)

print("Chi-square Statistic:", round(chi2_stat, 3))
print("P-value:", round(p_value, 4))
```

```python
'''26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results '''
# Simulate two datasets
data1 = np.random.normal(60, 10, 30)
data2 = np.random.normal(62, 15, 30)

# Variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

f_stat = var1 / var2
df1, df2 = len(data1) - 1, len(data2) - 1
p_value = stats.f.cdf(f_stat, df1, df2) if f_stat < 1 else 1 - stats.f.cdf(f_stat, df1, df2)

print("F-statistic:", round(f_stat, 3))
print("P-value:", round(p_value, 4))

# Visualize
plt.hist(data1, alpha=0.5, label="Data 1 (low var)")
plt.hist(data2, alpha=0.5, label="Data 2 (high var)")
plt.legend()
plt.title("F-test: Comparing Variances")
plt.grid(True)
plt.show()
```
